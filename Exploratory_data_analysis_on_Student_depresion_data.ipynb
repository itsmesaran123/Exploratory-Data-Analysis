## Introduction
#### Depression is a major public health issue that impacts people from different backgrounds and jobs. It is a complex condition shaped by mental, social, and environmental factors, often showing up as ongoing sadness, lack of interest, and reduced ability to function in everyday life. For students and workers, the stress of school performance, money issues, and balancing work and life can worsen mental health problems, making depression common in these groups.
#### This research looks at a dataset to understand what causes depression in students and professionals. The dataset contains various factors like demographic details, academic and job stress, lifestyle choices, and personal experiences. By studying these factors, the research aims to find patterns and connections that lead to depression, providing insights into key risk factors and possible ways to help.

## Obective
#### The study uses statistics and visual tools to find important links between depression and factors like money worries, school stress, sleep patterns, and personal experiences. The results are meant to help create ways to reduce depression, improve mental health, and shape support systems for students and workers.

### Data cleaning

#### renaming every columns that has whitespaces

```python
df = df.rename(columns={'Academic Pressure' : 'Academic_Pressure' ,'Work Pressure' : 'Work_Pressure','Study Satisfaction' : 'Study_Satisfaction' , 'Job Satisfaction' : 'Job_Satisfaction' , 'Sleep Duration': 'Sleep_Duration' ,'Dietary Habits' : 'Dietary_Habits' , 'Have you ever had suicidal thoughts ?' : 'Have_you_ever_had_suicidal_thoughts_?', 'Work/Study Hours' : 'Work/Study_Hours', 'Financial Stress' : 'Financial_Stress' , 'Family History of Mental Illness' : 'Family_History_of_Mental_Illness'})
df.isna().sum()
```

#### drop unnecesary columns

```python
clmn_to_drop =["Sleep_Duration"]
df.drop(columns=clmn_to_drop)
df.isna().sum()

```

#### changing data types

```python
df["id"].astype("int64")
df["Age"].astype("int64")
df["Academic_Pressure"].astype("int64")
df["Work_Pressure"].astype("int64")
df["CGPA"].astype("float64")
df["Study_Satisfaction"].astype("int64")
df["Job_Satisfaction"].astype("int64")
df["Work/Study_Hours"].astype("int64")
df["Financial_Stress"].astype("float64")
df["Depression"].astype("int64")

#df["City"].astype("string")
#df["Profession"].astype("string")
#df["Dietary_Habits"].astype("string")
#df["Degree"].astype("string")
#df["Have_you_ever_had_suicidal_thoughts_?"].astype("string")
#df["Family_History_of_Mental_Illness"].astype("string")
#df["Gender"].astype("string")
```

#### checking all columns for outliers

```python
df = df.select_dtypes(exclude=['object'])
for column in df:
    plt.figure(figsize=(17,1))
    sns.boxplot(data=df,x=column)
```

#### removing outliers

```python
df['z_score'] = (df['id'] - df['id'].mean()) / df['id'].std()
threshold=2
df = df[df["z_score"].abs() <= threshold]
df = df.drop(columns=["z_score"])
```

```python
df['z_score'] = (df['Age'] - df['Age'].mean()) / df['Age'].std()
threshold=1
df = df[df["z_score"].abs() <= threshold]
df = df.drop(columns=["z_score"])
```

```python
df['z_score'] = (df['Job_Satisfaction'] - df['Job_Satisfaction'].mean()) / df['Job_Satisfaction'].std()
threshold=1
df = df[df["z_score"].abs() <= threshold]
df = df.drop(columns=["z_score"])
```

```python
df['z_score'] = (df['CGPA'] - df['CGPA'].mean()) / df['CGPA'].std()
threshold=2
df = df[df["z_score"].abs() <= threshold]
df = df.drop(columns=["z_score"])
```

#### The study shows that depression in students and professionals is affected by various factors, including academic, work-related, lifestyle, and personal issues.
